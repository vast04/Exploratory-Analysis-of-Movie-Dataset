## Data Import and first Inspection

__Some additional information on Features/Columns__:

* **id:** The ID of the movie (clear/unique identifier).
* **title:** The Official Title of the movie.
* **tagline:** The tagline of the movie.
* **release_date:** Theatrical Release Date of the movie.
* **genres:** Genres associated with the movie.
* **belongs_to_collection:** Gives information on the movie series/franchise the particular film belongs to.
* **original_language:** The language in which the movie was originally shot in.
* **budget_musd:** The budget of the movie in million dollars.
* **revenue_musd:** The total revenue of the movie in million dollars.
* **production_companies:** Production companies involved with the making of the movie.
* **production_countries:** Countries where the movie was shot/produced in.
* **vote_count:** The number of votes by users, as counted by TMDB.
* **vote_average:** The average rating of the movie.
* **popularity:** The Popularity Score assigned by TMDB.
* **runtime:** The runtime of the movie in minutes.
* **overview:** A brief blurb of the movie.
* **spoken_languages:** Spoken languages in the film.
* **poster_path:** The URL of the poster image.
* **cast:** (Main) Actors appearing in the movie.
* **cast_size:** number of Actors appearing in the movie.
* **director:** Director of the movie.
* **crew_size:** Size of the film crew (incl. director, excl. actors).

## The best and the worst movies...

 __Filtering__ the Dataset and __finding the best/worst n Movies__ with the

- Highest Revenue
- Highest Budget
- Highest Profit (=Revenue - Budget)
- Lowest Profit (=Revenue - Budget)
- Highest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10) 
- Lowest Return on Investment (=Revenue / Budget) (only movies with Budget >= 10)
- Highest number of Votes
- Highest Rating (only movies with 10 or more Ratings)
- Lowest Rating (only movies with 10 or more Ratings)
- Highest Popularity

```python
from IPython.display import HTML

def bestworst(n, dataframe, by, ascending=False, min_bud=0, min_votes=0):
    df= dataframe.copy()
    df= df.rename(columns= {'budget_musd':'Budget', 'revenue_musd':'Revenue',
                           'vote_count': 'Votes', 'vote_average':'Average Rating'})
    df.Budget.fillna(0, inplace = True)
    df.Votes.fillna(0, inplace = True)
    df['Profit']= df['Revenue']- df['Budget']
    df['ROI']= df['Revenue']/df['Budget']
    
    df = df[(df['Budget']>=min_bud) & (df['Votes']>= min_votes)].sort_values(by, ascending=ascending).set_index('title')
    
    return HTML(df[['poster_path',by]].head(n).to_html(escape=False))              #(escape=False) because we want html to act like html
```

__Movies Top 5 - Highest Revenue__

```python
bestworst(5 ,df, 'Revenue')
```

__Movies Top 5 - Highest Budget__

```python
bestworst(5, df, 'Budget')
```

__Movies Top 5 - Highest Profit__

```python
bestworst(5, df, 'Profit')
```

__Movies Top 5 - Lowest Profit__

```python
bestworst(5, df, 'Profit', ascending=True)
```

__Movies Top 5 - Highest ROI__

```python
bestworst(5, df, 'ROI', min_bud=10)
```

__Movies Top 5 - Lowest ROI__

```python
bestworst(5, df, 'ROI', ascending= True, min_bud=10)
```

__Movies Top 5 - Most Votes__

```python
bestworst(5, df, 'Votes')
```

__Movies Top 5 - Highest Rating__

```python
bestworst(5, df, 'Average Rating', min_votes=10)
```

__Movies Top 5 - Lowest Rating__

```python
bestworst(5, df, 'Average Rating', ascending=True, min_votes=10)
```

__Movies Top 5 - Most Popular__

```python
bestworst(5, df, 'popularity')
```

## Are Franchises more successful?

- mean budget raised
- mean popularity
- mean rating

__Franchise vs. Stand-alone: Average Revenue__

## Most Successful Directors

6. __Finding__ the __most successful Directors__ in terms of

- __total number of movies__
- __total revenue__
- __mean rating__

```python
#total number of movies
df.groupby('director').size().sort_values(ascending=False)
```

```python
plt.figure(figsize=(8,6))
df.groupby('director').size().sort_values(ascending=False).head(20).plot(kind='bar')
plt.show()
```

```python
#total revenue
dir_rev= df.groupby('director').agg({'revenue_musd': 'sum'}).sort_values('revenue_musd', ascending=False).head(20)
dir_rev
```

```python
plt.figure(figsize=(8,6))
plt.bar(dir_rev.index, dir_rev['revenue_musd'])
plt.xticks(rotation= 90)
plt.xlabel('Director')
plt.ylabel('Total revenue (In MUSD)')
plt.show()
```

```python
#Mean Rating
mr= df.groupby('director').agg({'title':'count','vote_average':'mean', 'vote_count': 'sum'})
mean_rating= mr[(mr['title']>10) & (mr['vote_count']>5000)].sort_values('vote_average', ascending=False).head(20)
mean_rating
```

```python
plt.figure(figsize=(8,6))
plt.bar(mean_rating.index, mean_rating['vote_average'])
plt.xticks(rotation= 90)
plt.xlabel('Directors')
plt.ylabel('Mean Rating')
plt.show()
```

## Most Successful Actors

```python
actors_df=df.copy().set_index('id')
act= actors_df.cast.str.split('|', expand=True)
act=act.stack().reset_index(level=1, drop=True).to_frame()                    #stacking according to ids, dropping unnecessary index and converting to datframe.
```

```python
act= act.stack().reset_index(level=1, drop=True).to_frame() 
act.columns= ['Actor']
act
```

```python
actors_df= act.merge(df[['title', 'revenue_musd', 'vote_average', 'popularity']], 
                     how='left', left_index=True, right_index=True)
actors_df.rename(columns = {'vote_average':'Average_Rating'},inplace=True) 
actors_df
```

```python
# number of unique actors
actors_df['Actor'].nunique()
```

```python
#unique actors
actors_df['Actor'].unique()
```

```python
#actors with the most movies
actors_df.groupby('Actor').size().sort_values(ascending=False)
```

```python
plt.figure(figsize=(10,8))
actors_df.groupby('Actor').size().sort_values(ascending=False).head(20).plot(kind="bar")
plt.ylabel('Number of movies')
plt.show()
```

```python
act=actors_df.groupby('Actor').agg(Total_movies=('title','count'), Total_Revenue=('revenue_musd','sum'),
                                   Mean_Revenue= ('revenue_musd','mean'), Mean_Rating=('Average_Rating', 'mean'),
                                   Mean_popularity= ('popularity','mean'))
act
```

```python
# according to total revenue
act.sort_values('Total_Revenue', ascending=False).head(10)
```
